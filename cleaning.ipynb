```python
filtered_rows = Nasa[(Nasa['Death Mission'].isnull()) & (~Nasa['Death Date'].isnull())]

Nasa.drop(filtered_rows.index, inplace=True)
```

```python
Nasa['Birth Date'] = pd.to_datetime(Nasa['Birth Date'], dayfirst=True)
Nasa['Death Date'] = pd.to_datetime(Nasa['Death Date'], dayfirst=True)
```
